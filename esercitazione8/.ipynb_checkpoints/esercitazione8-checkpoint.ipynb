In meccanica quantistica, il principio variazionale afferma che il valore di aspettazione dell'hamiltoniana sullo stato fondamentale, costituisce un limite inferiore per la stessa quantià calcolata su un qualsiasi altro stato: 

$$
\langle {\hat H} \rangle_T = 
\frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)}
{\int dx |\Psi_T(x)|^2} \ge E_0 =
\frac{\langle \Psi_0| {\hat H} | \Psi_0 \rangle}
{\langle \Psi_0 | \Psi_0 \rangle}
$$

In questa esercitazione, si è ottimizzato tramite Variational Monte Carlo, il ground state di una particella in uno stazio 1D, confinata nel potenziale

$$
V(x) = x^4 - \frac{5}{2}x^2
$$

partendo da una funzione d'onda di prova, caratterizzata da un set di parametri variazionali $\mu$ e $\sigma$
$$
\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$$



### Esercizio 8.1

Come prima cosa si è implementato un codice che, tramite algoritmo di Metropolis, campiona $|\Psi_T^{\sigma,\mu}(x)|^2$, usando una distribuzione di probabilità di transizione $T(x_{new}|x_{old})$ uniforme, in modo da calcolare (con $\hbar = 1$ e $m=1$)

$$
\langle {\hat H} \rangle_T = 
\frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)}
{\int dx |\Psi_T(x)|^2} = 
\int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}
$$

necessaria all'ottimizzazione dei parametri $\mu$ e $\sigma$. 

### Esercizio 8.2

La ricerca dei parametri per minimizzare $\langle {\hat H} \rangle_T$ si è svolta tramite Simulated Annealing. Il sistema viene raffreddato lentamente partendo da una temperatura $T=2$ fino ad arrivare a $T=0.01$ e, in analogia con la termodinamica, si arriva alla configurazione in cui la cost function - nel nostro caso $\langle {\hat H} \rangle_T$ -, che svolge il ruolo dell'energia del sistema, è minima. 


```python
import numpy as np
import matplotlib.pyplot as plt

# Carica i dati saltando le prime 4 righe
T, H_SA, err_SA=np.loadtxt('esercizio8.2/H_SA_steps.dat', skiprows=1, usecols=(0,1,2), unpack=True)
blocks, H, err_H = np.loadtxt('esercizio8.2/dati.out', skiprows=4, usecols=(0,1,2), unpack=True)

fig, axs = plt.subplots(1, 2, figsize=(13, 4))
axs[0].errorbar(1/T, H_SA, yerr=err_SA, fmt='o', markersize=3, color='purple', ecolor='hotpink')

axs[0].set_xlabel(r'$\beta$')
axs[0].set_ylabel(r'$\langle H\rangle _{T}$')
axs[0].set_title(r'$<H>_{T}$ in funzione dei passi del Simulated Annealing')

axs[0].grid(True)

axs[1].errorbar(blocks, H, yerr=err_H, fmt='o', color='purple', ecolor='hotpink', capsize=3, label='Dati con errore')
axs[1].axhline(y=-0.46046587969259045, color='green', linestyle='--', linewidth=1.5, label='matrix equation')
axs[1].set_xlabel('blocks')
axs[1].set_ylabel(r'$\langle H\rangle _{T}$')
axs[1].set_title(r'$\langle H\rangle _{T}$ in funzione del numero di blocchi')
axs[1].legend()
axs[1].grid(True)
plt.show()
```

Il primo grafico mette in evidenza la correttezza dell'algoritmo di simulated annealing. Si osserva, infatti, che già dopo le prime decine di passi il valore di aspettazione dell'hamiltoniana diminuisce significativamente. 

Il secondo grafico, invece, mostra l'andamendo dell'hamiltoniana con i parametri ottimizzati $\mu= -0.728281$ e $\sigma=0.616058 $ in funzione del numero di blocchi usato nel data blocking per il calcolo della media. Il risultato è messo a confronto con quello che si otteiene scrivendo l'equazione di Schrodinger in forma matriciale (vedi cella successiva); i due si discostano a causa dei limiti del calcolo computazionale, così come accade per il plot di $|\Psi_T(x)|^2$, riportato di seguito. 

```python
import numpy as np
import matplotlib.pyplot as plt

def psi_mod2(x, mu, sigma):
    return (np.exp(-(x-mu)**2/(2*sigma**2))+np.exp(-(x+mu)**2/(2*sigma**2)))**2

filename = 'esercizio8.1/prob_psi.out'

data = np.loadtxt(filename)
mu=-0.728281
sigma=0.616058
x = data[:, 0]
y = data[:, 1]
x=np.sort(x)
psi_sqrd=psi_mod2(x, mu, sigma)

# Plot con barre di errore
plt.hist(x, 50, density=True, color='plum', label='sampled', edgecolor='black')
plt.plot(x, psi_sqrd/(np.sqrt(np.pi)*4*sigma**2), color='navy', label='analitic')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot dei dati con errore su y')
plt.grid(True)


def Vpot(x):
    return (x**2 - 2.5)*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

print("Ground state energy: ", E[0])

# Plot a few things
scale = 0.3
plt.plot(x,(psi[0])**2, label='matrix equation')
plt.title("Potential & Probabilities")
plt.xlabel("x")
plt.grid(True)
plt.xlim((-3,3))
plt.ylim((0,0.5))
plt.legend()
plt.show()
```
