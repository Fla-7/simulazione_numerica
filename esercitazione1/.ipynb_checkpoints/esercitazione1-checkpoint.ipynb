3. Si Ã¨ verificata l'ipotesi che i numeri non siano estratti da una distrbuzione uniforme con il test del $\chi^2$.

Dividendo l'intervallo $[0,1]$ in $M$ sottointervalli e facendo n lanci, ci si aspetta di osservare $n/M$ eventi in ogni sottointervallo. 
L'esercizio si Ã¨ svolto con $M=100$, ripetendo il test del $\chi^2$ 100 volte con $10000$ numeri estratti, calcolando 
$$\chi^2 = \sum_{i=1}^M \frac{\left( n_i - n/M \right)^2}{n/M}$$


```python
from scipy.stats import chi2

x=np.loadtxt("esercizio1.1.2_3/chiquadro.out", usecols=(0))
chi=np.loadtxt("esercizio1.1.2_3/chiquadro.out", usecols=(1))

fig, (F1, F2) = plt.subplots(1, 2, figsize=(14, 5))


F1.errorbar(x,chi,marker='o', color='hotpink', markersize='4')
F1.set_xlabel('#throws')
F1.set_ylabel(r'$\chi^2$')
F1.grid()

F2.hist(chi, bins=15, density=True, color='hotpink', edgecolor='black', alpha=0.7) 
x_teor = np.linspace(min(chi), max(chi), 100-1)
F2.plot(x_teor, chi2.pdf(x_teor, df=99), 'r-', label=r'Distribuzione $\chi^2$ (gradi di libertÃ =99)')

F2.set_title('Istogramma della Distribuzione dei Valori')
F2.set_xlabel('Valore')
F2.set_ylabel('Frequenza (Conteggio)')
F2.legend()

plt.grid(axis='y', alpha=0.5)
plt.show()
```

## Esercizio 1.2
- Si Ã¨ esteso il generatore di numeri pseudo-casuali per verificare il Teorema del Limite Centrale.

Si sono aggiunte due distribuzioni di probabilitÃ  tramite il metodo di inversione della distribuzione cumulativa per estrarre da una distribuzione 
esponenziale $p(x) = \lambda \exp(-\lambda x)$, $x\in [0;+\infty]$ e da una distribuzione di Cauchy-Lorentz $p(x)=\frac{1}{\pi}\frac{\Gamma}{(x-\mu)^2+\Gamma^2}$, $x\in [-\infty;+\infty]$. 

```python
from scipy.stats import norm, cauchy
file_names=['esercizio1.2/distribuzioni_N1.dat', 'esercizio1.2/distribuzioni_N2.dat', 'esercizio1.2/distribuzioni_N10.dat', 'esercizio1.2/distribuzioni_N100.dat']
pink_palette = ['#FF1493',  # DeepPink
                '#FF69B4',  # HotPink
                '#DB7093',
                '#FFB6C1']
titles=['dado standard', 'dado esponenziale', 'dado lorentziano']
N=[1,2,10,100]
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(15, 12))
cols=np.zeros(3)


for idx, filename in enumerate(file_names):
    unif=np.loadtxt(file_names[idx], usecols=(0)) #uniforme
    exp=np.loadtxt(file_names[idx], usecols=(1)) #esponenziale
    lor=np.loadtxt(file_names[idx], usecols=(2)) #lorenziana

    cols=[unif, exp, lor]
    
    for j in range(3):
        ax=axes[idx,j]
        current_color=pink_palette[idx]
        if j==2:
            bin=np.linspace(-25, 25, 50)
        else:
            bin=50
        ax.hist(cols[j], bins=bin, density=True, color=current_color, edgecolor='black', alpha=0.7)
        ax.set_title(f'{titles[j]} - N={N[idx]}', fontsize=10)

        if idx==3 and j!=2:
            mu_calc, sigma_calc=norm.fit(cols[j])
            
            #Generazione dei punti per la Gaussiana teorica
            xmin, xmax = ax.get_xlim()
            x = np.linspace(xmin, xmax, 100)
            
            # Calcolo della PDF Normale (Gaussiana) con i parametri calcolati
            p = norm.pdf(x, mu_calc, sigma_calc)
            
            # 5. Plottiamo la curva sovrapposta
            ax.plot(x, p, 'r-', linewidth=2, label=r'Gaussiana Teorica')
            ax.legend(fontsize=8)
        elif idx==3 and j==2: #Cauchy-Lorentz
            
                mu_lorentz, gamma_lorentz= cauchy.fit(cols[j])
                
                # Protezione da divisione per zero nel caso estremo, anche se improbabile
                if gamma_lorentz == 0:
                    gamma_lorentz = 1e-6 

                # Genera e plotta la curva Cauchy-Lorentz
                xmin, xmax = ax.get_xlim()
                x = np.linspace(xmin, xmax, 200) # PiÃ¹ punti per una curva Cauchy-Lorentz
                
                # Usa la PDF della distribuzione cauchy
                p = cauchy.pdf(x, loc=mu_lorentz, scale=gamma_lorentz)
                
                ax.plot(x, p, 'g-', linewidth=2, label=r'Cauchy-Lorentz Teorica')
                ax.legend(fontsize=8)
            
        

plt.suptitle('Confronto delle Medie Campionarie ', fontsize=16, y=1.02)
plt.tight_layout(rect=[0, 0, 1, 1.01]) # Adatta il layout includendo il suptitle
plt.show()
```

L'immagine contiene gli istogrammi per $10^4$ valori di $S_N = \frac{1}{N}\sum_{i=1}^N x_i$ (for $N=1, 2, 10, 100$), dove $x_i$ Ã¨ estratta con un dado standard, esponenziale ($\lambda=1$) e lorentziano ($\mu=0$ e $\Gamma=1$). 

Si osserva il compimento del Teorema del Limite Centrale nel caso di $N=100$; possiamo, infatti, fittare la distribuzione uniforme e quella esponenziale con una gaussiana e la Lorentziana con una distribuzione di Cauchy-Lorentz. 

### Esercizio 1.3
- Si Ã¨ simulato l'esperimento di Buffon per la stima di $\pi$.

La probabilitÃ  $P$ che l'ago intersechi una delle linee Ã¨ $P = 2L/\pi d$. Se l'ago viene lanciato $N_{thr}$ volte e colpisce la riga $N_{hit}$ di queste volte, si puÃ² stimare $\pi$ come $$\pi = \frac{2L}{Pd} = \lim_{N_{thr} \to \infty}\frac{2LN_{thr}}{N_{hit}d}$$


```python
x=np.loadtxt("esercizio1.3/dati1.3.out", usecols=(0))
ave=np.loadtxt("esercizio1.3/dati1.3.out", usecols=(1))
err=np.loadtxt("esercizio1.3/dati1.3.out", usecols=(2))

x*=100 # Number of throws = block * (Number of throws in each block)
plt.errorbar(x,ave,yerr=err,color='hotpink')
plt.xlabel('#throws')
plt.ylabel(r'$\pi$')
plt.grid(True)
plt.show()
```

IL grafico mostra la stima di $\pi$  con la sua incertezza (deviazione standard della media) per un grande numero di lanci $M$ ($M\ge 10^4$) in funzione del numero di blocchi $N$</font>. Si osserva che il risultato ottenuto converge al valore corretto di $\pi$ entro una deviazione standard della media.
