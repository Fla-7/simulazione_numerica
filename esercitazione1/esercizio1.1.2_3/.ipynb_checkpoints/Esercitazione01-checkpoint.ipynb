## Esercizio 1.1
Fare un grafico della stima di $\langle r \rangle=\int_0^1 r dr=1/2$ e della sua incertezza (che corrisponde alla Deviazione Standard per la stima di di $\langle r \rangle$) con un grande numero di *throw* $M$ (ad esempio $M=10^4$), in funzione del numero di blocchi $N$. 


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

M=10000             # numero totale di tiri
N=100                 # numero di blocchi
L=int(M/N)            # numero di tiri in ogni blocco
ave=np.zeros(N)
sigma=np.zeros(N)
x = np.arange(N)+1

ave=np.loadtxt("mediablocchi.out")
sigma=np.loadtxt("incertezze.out")

x*=L # Number of throws = block * (Number of throws in each block)
plt.errorbar(x,ave-0.5,yerr=sigma)
plt.xlabel('#throws')
plt.ylabel('I-0.5')
plt.grid(True)
plt.show()
    
```
