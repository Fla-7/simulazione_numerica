### Il problema del commesso viaggiatore

Si consideri un commesso viaggiatore che deve visiatre $N$ città con posizioni $\vec{c}_i$ prima di ritornare alla città di partenza. L'obbiettivo di questa esercitazione è ottimizzare il percorso, trovando l'ordine in cui le città vanno visitate in modo che venga percorsa la traiettoria pià corta possibile. Si deve perciò implementare un algoritmo di ottimizzazione dove la loss function è 

$$
L^{(2)}(\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N) = \sum_{i=1}^{N} |\vec{x}_i-\vec{x}_{i+1}|^2
$$

oppure
$$
L^{(1)}(\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N) = \sum_{i=1}^{N} |\vec{x}_i-\vec{x}_{i+1}|
$$


(con $\vec{x}_{N+1}=\vec{x}_1$).


### Esercizio 9.1

Si è implementato un algoritmo genetico così organizzato:

- classe City: contiene un vector di double che rappresentano le coordinate delle città, estratte randomicamente;
- classe Chromosome: contiene un vector di City, costituisce quindi un percorso possibile;
- classe Population: contiene un vector di Chromosome, è un insieme di percorsi.

Dopo aver estretto le coordinate per 34 città, o sulla circonferenza o nel quadrato, si genera una popolazione di N individui attraverso la permutazione di coppie di città; la popolazione viene ordinata in modo che il percorso più breve sia il primo. A questo punto avveiene la riproduzione, tramite crossover, con una probabilità del 70%: entrambi i percorsi scelti come genitori, attraverso la funzione selection che estrae un indice $j= int(M \times r^p)+1$, vengono tagliati nello stesso punto, scelto casualmente; mentre la prima parte di ogni percorso rimane fissa, la seconda viene scambiata reciprocamente. 

I figli così creati hanno il 10% di prbabilità di mutare con le funzioni:

- shift: m città vicine vengono spostate di n posizioni
- permutation: m città vicine vengono scambiate con altre m città vicine
- inversion: si inverte l'ordine in cui compaiono m città vicine

I nuovi individui vengono ordinati e il processo di riproduzione e mutazione si ripete. 

Per tutto lo svolgimento del codice vanno rispettati due vincoli: la prima città deve rimanere fissa e ogni città deve essere visitata una e una sola volta. 

Raggiunto un numero di generazioni sufficiente, il primo individuo della popolazione dovrebbe essere il migliore.
