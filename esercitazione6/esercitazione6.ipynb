Il modello di Ising consiste nel considerare un reticolo di spin la cui interazione è causa del magnetismo nella materia. In meccanica statistica il modello ha una soluzione esatta nel caso unodimensionale. L'hamiltoniana per un sistema di N spin che interagiscono con i primi vicini è (considerando $\mu_B=1$ and $k_B=1$):

$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) 
$$

dove $\{ s_i \}_{i=1,N}$ sono varibili di spin che possono assumere i valori $\{-1;1\}$, $J>0$ (modello ferromagnetico), e con l'assunzione che $s_{N+1} = s_{1}$.

La funzione di partizione si ricava essere

$$
Z=\sum_{\{ s_i \}} \exp \left( -\beta H\right) = \lambda_1^N + \lambda_2^N
$$

where
$\lambda_{1,2} = 
e^{\beta J} 
\cosh (\beta h) \pm \left[ e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J) \right]^{1/2}$, 
and where $\beta=\frac{1}{k_b T} = \frac{1}{T}$.

Da questa si possono ricavare le quantità termodinamiche di interesse. 

### Esercizio 6.1
In questo esercizio si è implementata la simulazione di un modello di Ising 1D tramite due diversi algoritmi di campionamento per il peso di Boltmann $p(\{s_i\})=\frac{e^{-\beta H}}{Z}$:
- algoritmo di Metropolis

    ogni mossa proposta (flip di spin) viene accettata con probabilità $\alpha=min[1, \frac{p(y)}{p(x)}]$, dove $p(y)$ è il peso di Boltzmann     della nuova configurazione proposta e $p(x)$ quello della configurazione attuale;
  
- algoritmo di Gibbs

    le mosse vengono proposte a partire dalla probabilità condizionata, pertanto la probabilità di transizione diventa $p(s \rightarrow s')=\frac{1}{1+e^{\pm2 \beta J \sum_{i} s_{i}}}$ e le mosse sono sempre accettate

Si è quindi aggiunto il calcolo di:
1. $U(N,T) = \langle H \rangle_{h=0}$
2. $C(N,T) = k_B \beta^2 (\langle H^2 \rangle_{h=0}-\langle H \rangle_{h=0}^2)$ 
3. $\chi(N,T) = \beta \langle (\sum_{i=1,N}s_i)^2 \rangle_{h=0}$
4. $M(N,T,h=0.02)= \langle \sum_{i=1,N}s_i \rangle_{h=0.02}$

Si riportano di seguito i risultati della simulazione svolta con data blocking - 200 blocchi da 20000 passi. 

```python
import numpy as np
import matplotlib.pyplot as plt

#teorica
points=100
T = np.linspace(0.2,3.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
x=np.zeros(11)

dati_metro=np.loadtxt('simulazioni_eseguite/Metropolis/total_energy.dat')
temp=dati_metro[:,0]
TE_metro=dati_metro[:,1]
err_metro=dati_metro[:,2]

dati_gibbs=np.loadtxt('simulazioni_eseguite/Gibbs/total_energy.dat')
TE_gibbs=dati_gibbs[:,1]
err_gibbs=dati_gibbs[:,2]

fig , axes =plt.subplots(1, 2, figsize=(12, 5))

#fig.set_title('Ising 1D, internal energy')

axes[0].plot(T, e, label='exact curve')
axes[0].errorbar(temp, TE_metro, yerr=err_metro, marker='o', markersize=3, color='purple', ecolor='hotpink', linestyle='none', label='Metropolis')
axes[0].set_title('internal energy - Metropolis')
axes[0].set_xlabel('T')
axes[0].set_ylabel('U/N')
axes[0].grid()
axes[0].legend()

axes[1].plot(T, e, label='exact curve')
axes[1].errorbar(temp, TE_gibbs, yerr=err_gibbs, marker='o', markersize=3, color='purple', ecolor='hotpink', linestyle='none', label='Gibbs')
axes[1].set_title('internal energy - Gibbs')
axes[1].set_xlabel('T')
axes[1].set_ylabel('U/N')
axes[1].grid()
axes[1].legend()

plt.show()
```

```python
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)

dati_metro=np.loadtxt('simulazioni_eseguite/Metropolis/specific_heat.dat')
temp=dati_metro[:,0]
C_metro=dati_metro[:,1]
err_metro=dati_metro[:,2]

dati_gibbs=np.loadtxt('simulazioni_eseguite/Gibbs/specific_heat.dat')
C_gibbs=dati_gibbs[:,1]
err_gibbs=dati_gibbs[:,2]

fig , axes =plt.subplots(1, 2, figsize=(12, 5))

axes[0].plot(T, heat, label='exact curve')
axes[0].errorbar(temp, C_metro, yerr=err_metro, marker='o', markersize=3, color='purple', ecolor='hotpink', linestyle='none', label='Metropolis')
axes[0].set_title('heat capacity - Metropolis')
axes[0].set_xlabel('T')
axes[0].set_ylabel('C')
axes[0].grid()
axes[0].legend()

axes[1].plot(T, heat, label='exact curve')
axes[1].errorbar(temp, C_gibbs, yerr=err_gibbs, marker='o', markersize=3, color='purple', ecolor='hotpink', linestyle='none', label='Gibbs')
axes[1].set_title('heat capacity - Gibbs')
axes[1].set_xlabel('T')
axes[1].set_ylabel('C')
axes[1].grid()
axes[1].legend()

plt.show()
```

```python
h=0.02 #external field
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)

dati_metro=np.loadtxt('simulazioni_eseguite/Metropolis/magnetization.dat')
temp=dati_metro[:,0]
M_metro=dati_metro[:,1]
err_metro=dati_metro[:,2]

dati_gibbs=np.loadtxt('simulazioni_eseguite/Gibbs/magnetization.dat')
M_gibbs=dati_gibbs[:,1]
err_gibbs=dati_gibbs[:,2]

fig , axes =plt.subplots(1, 2, figsize=(12, 5))

axes[0].plot(T, M, label='exact curve')
axes[0].errorbar(temp, M_metro, yerr=err_metro, marker='o', markersize=3, color='purple', ecolor='hotpink', linestyle='none', label='Metropolis')
axes[0].set_title('magnetization - Metropolis')
axes[0].set_xlabel('T')
axes[0].set_ylabel(r'$M$')
axes[0].grid()
axes[0].legend()

axes[1].plot(T, M, label='exact curve')
axes[1].errorbar(temp, M_gibbs, yerr=err_gibbs, marker='o', markersize=3, color='purple', ecolor='hotpink', linestyle='none', label='Gibbs')
axes[1].set_title('magnetization - Gibbs')
axes[1].set_xlabel('T')
axes[1].set_ylabel(r'$M$')
axes[1].grid()
axes[1].legend()

plt.show()
```

```python
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)

dati_metro=np.loadtxt('simulazioni_eseguite/Metropolis/susceptibility.dat')
temp=dati_metro[:,0]
X_metro=dati_metro[:,1]
err_metro=dati_metro[:,2]

dati_gibbs=np.loadtxt('simulazioni_eseguite/Gibbs/susceptibility.dat')
X_gibbs=dati_gibbs[:,1]
err_gibbs=dati_gibbs[:,2]

fig , axes =plt.subplots(1, 2, figsize=(12, 5))

axes[0].plot(T, X, label='exact curve')
axes[0].errorbar(temp, X_metro, yerr=err_metro, marker='o', markersize=3, color='purple', ecolor='hotpink', linestyle='none', label='Metropolis')
axes[0].set_title('susceptibility - Metropolis')
axes[0].set_xlabel('T')
axes[0].set_ylabel(r'$\chi$')
axes[0].grid()
axes[0].legend()

axes[1].plot(T, X, label='exact curve')
axes[1].errorbar(temp, X_gibbs, yerr=err_gibbs, marker='o', markersize=3, color='purple', ecolor='hotpink', linestyle='none', label='Gibbs')
axes[1].set_title('susceptibility - Gibbs')
axes[1].set_xlabel('T')
axes[1].set_ylabel(r'$\chi$')
axes[1].grid()
axes[1].legend()

plt.show()
```
